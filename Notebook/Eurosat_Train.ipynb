# Training Performance

```python
# In terms of accuracy
acc_train = np.load('acc_train.npy')
acc_val = np.load('acc_val.npy')
```

```python
# plotting
plt.title("Accuracy while training")
plt.xlabel("Epochs")
plt.ylabel("Training Accuracy")
plt.plot(range(len(acc_train)), acc_train, color ="red")
plt.plot(range(len(acc_train)), acc_val, color ="blue")
plt.show()
```

```python
# In terms of loss
loss_train = np.load('loss_train.npy')
```

```python
# plotting
plt.title("loss while training")
plt.xlabel("Epochs")
plt.ylabel("Training loss")
plt.plot(range(len(loss_train)), loss_train, color ="red")
plt.show()
```
